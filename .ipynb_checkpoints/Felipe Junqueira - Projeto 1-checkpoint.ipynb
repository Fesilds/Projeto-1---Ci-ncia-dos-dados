# Projeto 1: Ciência dos dados


Conflitos políticos em cima do bloco economico Mercosul tem tomado as midias a pouco tempo, sendo matéria de reportagens na mídia como no G1 "Mercosul discute reformas internas para enfrentar concorrência europeia" e em o Globo "Ao assumir comando de Mercosul, Brasil promete fazer revisão completa de taxas de importação". Esse projeto pretende avaliar a efetividade de determinados blocos economicos em paises membros, esse projeto procura detrminar a funcionalidade do bloco economico "Mercosul" (1991 - ) nos paises membros a partir de uma analise crítica de diferentes dados revelevantes de um pais, como: PIB per cap; IDH; Desemprego (25-50 anos), comparando com outros paises membros de outros blocos
economicos e não membros, para assim ter uma visao ampliada da utilidade do bloco em um determinado pais. 


### Perguntas


- Após se juntar a mercosul o pib do pais deve crescer consideravelmente?

- Apos um dado tempo da participação de um pais no Mercosul seu IDH deve aumentar?

- Participando de um bloco economico a taxa de desemprego diminue considerávelmente?

### Hipóteses:

- Se juntando a um bloco economico como o Mercosul, o pais deve aumentar sua produção interna para expotação, uma vez que o bloco economico facilita relações comerciais entre pares.

- Com o aumento na intensidade do comercio nos paises do bloco, como uma cosequencia passiva o IDH deve aumentar mais do que cresceria sem o auxilio do bloco economico.

- Com um aumento no comerico entre os paises pertencentes ao bloco econoico, deve causar um aumento na produção resultando em criações de trabalhos reduzindo o desemprego.

# Tabela Desemeprego IDH

```python
c = a.join(b,how="inner",lsuffix=" desemprego",rsuffix=" IDH") 
c
```

```python
def reta(dados_x, dados_y, c):    
    a = dados_y.cov(dados_x) / dados_x.var()
    b = dados_y.mean() - a*dados_x.mean()

    print('Ajuste linear entre X3 e X4:')
    print('Coeficiente linear:', b)
    print('Coeficiente angular:', a)
    plt.scatter(dados_x, dados_y, c=c, alpha=0.8)

    plt.plot((dados_x.min(), dados_x.max()), (a*dados_x.min()+b, a*dados_x.max()+b), color=c)

    plt.tight_layout()
```

```python

plt.rcParams["axes.titlesize"] = 25
plt.rcParams["xtick.labelsize"] = 15
plt.rcParams["ytick.labelsize"] = 15
plt.rcParams["axes.labelsize"] = 17
plt.rcParams["lines.markersize"] = 15

plt.scatter(c['2000 desemprego'],c['2000 IDH'], color="y")

plt.scatter(c['2010 desemprego'],c['2010 IDH'], color="g" )
plt.xlabel("Desemprego")
plt.ylabel("IDH")
plt.title("IDH X Desemprego")
reta(c['1991 desemprego'],c['1991 IDH'],"r")
```

No grafico a cima IDH Desmeprego é possível analisar um aumento no idh e uma redução no desmeprego, com uma reta de coeficiente angular positivo mostrando que os possiveis resultados para o futuro esperam valores que traduzem um bom indice de idh com um desemprego pequeno.
